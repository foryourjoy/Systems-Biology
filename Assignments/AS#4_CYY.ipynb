```python
import numpy as np #import numpy
import random #import random
import matplotlib.pyplot as plt #import matplotlib for graphs
```

# 1) Generate an observed data [y] containing noise using
# y=3x+2 where x = {1,2,3,4,5,6,7,8,9}

```python
np.random.seed(10) #create random numbers
x = np.arange(1,10) #1~9 range of x
y = 3*x+2 #function of x
noise = np.random.normal(0,1,len(y)) #make noise
observed_y = y+noise #function of x with some noise
print(f"x={x}")
print(f"observed_y={observed_y}")
```

# 2) Plot the true data and the observed data

```python
plt.plot(x,y) #real function
plt.scatter(x,observed_y,color="black") #observed function
```

# Case1. 'a' is unknown

# 3) Randomly assign the parameter values 'a' N times

```python
N=10 # set the number of estimation trials
np.random.seed(10) # make random data reproducible
a = np.random.uniform(size=N, low=0, high=10) # create a random vector
plt.figure(figsize = (15,15))

#making subgraph
for i in range(1,N):
    x=np.arange(1,10)
    y=a[i]*x+2
    MSE = np.square(np.subtract(observed_y, y)).mean() #calculate MSE
    plt.subplot(3,3,i)
    plt.scatter(x,observed_y,color="black") #function of x with noise
    plt.ylim(0,40) #y axis limitation
    plt.text(2,35,f"y = {a[i].round(2)}x + 2") #write function detail on the graph
    plt.text(2,33,f"MSE = {MSE.round(2)}") #write MSE on the graph
    plt.plot(x,y) #x and y with parameter a
```

#5) Let the code find the minimum loss and the corresponding parameter values

```python
N=10 # set the number of estimation trials
np.random.seed(10) # make random data reproducible
a = np.random.uniform(size=N, low=0, high=10) # create a random vector
MSE=[] #list of MSE

for i in range(0,N): 
    x=np.arange(1,10)
    y=a[i]*x+2 #random y with parameter a
    MSE.append(np.square(np.subtract(observed_y, y)).mean()) #append MSE values

minimum=min(MSE) #find minimum value of MSE
min_x=MSE.index(minimum) #find the index of MSE
print(a,MSE)
#plot
plt.plot(np.arange(1,N+1),MSE,marker='o',color='black')
plt.axvline(min_x+1, 0, 1000, color='red', linestyle='--')
plt.plot (min_x+1,minimum,'ro')
plt.ylabel('Loss: MSE')
plt.xlabel('Number of trials')
plt.yticks(np.arange(0,850,200))
plt.title('Parameter estimation: N=10 trials')
```

# 6) Let the code to plot the observed data and the output of the best model together

```python
N=10 # set the number of estimation trials
np.random.seed(10) # make random data reproducible
a = np.random.uniform(size=N, low=0, high=10) # create a random vector
a_sorted = np.sort(a) #sort by ascending order
MSE=[]
for i in range(0,N):
    x=np.arange(1,10)
    y=a_sorted[i]*x+2 #y values with sorted a
    MSE.append(np.square(np.subtract(observed_y, y)).mean()) #append MSE values
minimum=min(MSE)
min_x=MSE.index(minimum)

y=a_sorted[min_x]*x+2 #find the predicted function
plt.plot(x,y)
plt.scatter(x,observed_y,color="black")
plt.ylim(0,40) #y axis limitation
plt.text(2,35,f"y = {a_sorted[min_x].round(2)}x + 2") #write function detail on the graph
plt.text(2,33,f"MSE = {MSE[min_x].round(2)}") #write MSE on the graph
plt.yticks(np.array([-1,0,10,20,30,40,41]),labels=['',0,10,20,30,40,''])
plt.xticks(np.array([-0.2,0,2,4,6,8,10,10.2]),labels=['',0,2,4,6,8,10,''])
plt.xlabel('time(hr)')
plt.ylabel('[X]')
plt.title('The best model found with 10 trials')
```
