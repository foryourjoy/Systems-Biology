```python
plt.plot(as1['time'],as1['x'],label='x') #graph of x
plt.plot(as1['time'],as1['y'],label='y') #graph of y
plt.plot(as1['time'],as1['z'],label='z') #fraph of z

plt.ylim(0.3,1.3) #range of y axis

#label and title
plt.xlabel('Time')
plt.ylabel('Concentrations')
plt.title('A dynamic behavior of a system')
plt.legend(loc='upper right')

plt.show()
```
